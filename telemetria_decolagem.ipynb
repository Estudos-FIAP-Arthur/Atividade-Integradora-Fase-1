# Sistema de Verificação de Decolagem

```python

def verificar_decolagem():
    temperatura_interna = float(input("Temperatura interna (°C): "))
    temperatura_externa = float(input("Temperatura externa (°C): "))
    integridade = int(input("Integridade estrutural (0 ou 1): "))
    energia = float(input("Energia (%): "))
    pressao = float(input("Pressão dos tanques (psi): "))

    navegacao = int(input("Navegação (0 ou 1): "))
    comunicacao = int(input("Comunicação (0 ou 1): "))
    propulsao = int(input("Propulsão (0 ou 1): "))
    termico = int(input("Controle térmico (0 ou 1): "))

    if integridade == 0:
        print("DECOLAGEM ABORTADA")
        return

    if temperatura_interna < 18 or temperatura_interna > 27:
        print("DECOLAGEM ABORTADA")
        return

    if temperatura_externa < -20 or temperatura_externa > 40:
        print("DECOLAGEM ABORTADA")
        return

    if energia < 80:
        print("DECOLAGEM ABORTADA")
        return

    if pressao < 70 or pressao > 120:
        print("DECOLAGEM ABORTADA")
        return

    if (navegacao == 0 or comunicacao == 0
        or propulsao == 0 or termico == 0):
        print("DECOLAGEM ABORTADA")
        return

    print("PRONTO PARA DECOLAR")

verificar_decolagem()
```
